## This is a script to plot mass in solar mass unit against time in Kyr using FUNCTIONS for runs that have more than nine sinks

```python
import numpy as np
import math
from astropy import constants as const
from astropy import units as u
```

```python
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
timeUnitSi = (u.m**3/const.G.unit/u.kg)**(0.5); timeUnitSi
```

```python
timeUnit = (const.pc**3/(const.G)/const.M_sun)**(0.5); timeUnit
```

```python
def openfile (sinkFilePath):
    print("Opening " + sinkFilePath)
    time, mass = np.loadtxt(sinkFilePath,unpack=True, usecols=(0,4))
    time=time*timeUnit # time in second
    time = time/u.yr.to('s')/1000 # time in kyr
#     time=time/(31.5576*10**9)
    return time, mass
```

```python
def plotsink(time, mass):
#     fig1.add_subplot(1,1,1)
    fontSize = 28
    plt.ylabel('core mass')
    plt.xlabel('time(kyr)')
    Label = "sinks"
#     ax1 = plt.subplot(1,1,1)
#     if i == 1:
#         print(i)
    ax1.plot(time, mass, label = Label, linestyle = "--", color="k")
#     else:
#         print(i)
#         sinkMarker = "s"
#         ax1.plot(time, mass, label = Label, linestyle = "-", linewidth ="5" ,color="b", alpha=0.4)
#     leg = plt.legend(loc='best') ## Add a legend at the best location
```
